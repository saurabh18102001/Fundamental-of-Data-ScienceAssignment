```python
#7. Transform project column by reducing category variations to common names.
mapping = {
    'Strong': 1,
    'Average': 2,
    'Weak': 3,
}

df[['Python', 'Java', 'SQL']] = df[['Python', 'Java', 'SQL']].apply(lambda x: x.map(mapping))
```
