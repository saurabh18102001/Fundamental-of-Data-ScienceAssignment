```python
player_list = list(df["player_of_match"].value_counts()[0:5].keys())
player_list
```

```python
player_match = list(df["player_of_match"].value_counts()[0:5].values)
player_match
```

```python
from matplotlib import pyplot as pl
pl.bar(player_list, player_match,color = ["r","y","orange","g","blue"])
pl.show()
```

```python
df["result"].value_counts()
result = list(df["result"].value_counts().keys())
result_type = list(df["result"].value_counts().values)
pl.bar(result,result_type)
pl.show()
```
