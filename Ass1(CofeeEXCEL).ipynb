4. Create new columns for day, month and year

```python
df['transaction_date'] = pd.to_datetime(df['transaction_date'])
```

```python
df['day'] = df['transaction_date'].dt.day
df['month'] = df['transaction_date'].dt.month
df['year'] = df['transaction_date'].dt.year
df.head()
```

5. In cofFee-shop dataset find out sales for each month

```python
df.groupby('month')['unit_price'].sum()
```

6. In cofFee-shop dataset find out total sales for each product category.

```python
df.groupby('product_category')['unit_price'].sum()
```

7. Find out minimum and maximum price values and name of the product.

```python
df.loc[df['unit_price'] == df['unit_price'].max()][['product_category','unit_price']]
```

```python
df.loc[df['unit_price'] == df['unit_price'].min()][['product_category','unit_price']]
```
