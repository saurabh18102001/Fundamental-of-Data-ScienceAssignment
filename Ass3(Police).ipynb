Fetch Stop Hour From the column of stop time.

```python
#2. Fetch Stop Hour From the column of stop time.
df = police['stop_time'].str.split(':', expand=True)
stop_hour = df[0]
print(stop_hour.head())
```

```python
#3. Insert new column named stop_daytime and create four slots of Morning, Afternoon, Evening and Nigth based on hour ranges 6 to 11 AM,
#11 AM Â to 4 PM, 5PM to 9 PM and 10 PM to 5 AM.
police['stop_hour'] = police['stop_time'].str.split(':', expand=True)[0].astype(int)

def categorize_time(hour):
    if 6 <= hour <= 11:
        return 'Morning'
    elif 12 <= hour <= 16:
        return 'Afternoon'
    elif 17 <= hour <= 21:
        return 'Evening'
    else:
        return 'Night'

police['stop_daytime'] = police['stop_hour'].apply(categorize_time)
police.head()
```
